```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
```

```python
df_jan=pd.read_parquet('yellow_tripdata_2023-01.parquet')
```

```python
df_jan.head(5)
```

```python
df_jan.info()
```

```python
df_jan['duration']= df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']
```

```python
df_jan.duration=df_jan.duration.apply(lambda td: td.total_seconds()/60)
```

```python
df_jan.head(10)
```

```python
df_jan.duration.describe(percentiles=[0.90, 0.92, 0.95, 0.98])
```

```python
((df_jan.duration>=1) & (df_jan.duration<=60)).mean()
```

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
df_jan[categorical] = df_jan[categorical].astype(str)
```

```python
train_dicts = df_jan[categorical + numerical].to_dict(orient='records')
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
train_dicts
```

```python
dv.feature_names_
```

```python
target = 'duration'
y_train = df_jan[target].values
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, y_pred, squared=False)
```
